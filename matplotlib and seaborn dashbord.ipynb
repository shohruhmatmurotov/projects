```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv(r"C:\\Users\\Asus\\Downloads\\customer_orders.csv")
```

```python
chart=data.groupby('Product').agg({"Quantity":'sum','Price':'sum'}).reset_index().sort_values('Price',ascending=False).head(5)
chart
```

```python
custom_palette = ['#FF9999', '#66B3FF', '#99FF99', '#FFCC99']
plt.figure(figsize=(6,4),dpi=300)

sns.barplot(data=chart ,x='Product' ,y='Price' ,hue='Quantity' ,edgecolor='black',palette=custom_palette)
plt.ylabel('The overall sales price per product')
plt.xlabel('The most expensive products ')

plt.legend()
plt.show()
```

```python
graph=data.groupby(['CustomerID']).agg({'Price':'mean'}).reset_index().sort_values('Price',ascending=False).astype(int)
graph
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 

plt.figure(figsize=(6,5) ,dpi=300)
sns.lineplot(data=graph, x='CustomerID', y='Price')
plt.xticks([101,102,103,104,105])
plt.xlabel('The customers')
plt.ylabel('The avarage selling price per customers')
plt.show()
sns.set_style('whitegrid')


```
